```python
# Plotting the total number of responses
response_counts = raw_data['Response'].value_counts()
response_counts.plot(kind='bar')
plt.title('Total Number of Responses')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate response rate by sales channel
response_by_channel = raw_data.groupby('Sales Channel')['Response'].value_counts().unstack() #normalize=True
response_by_channel.plot(kind='bar', stacked=True)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.legend(title='Response', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# Calculate response rate by sales channel normalized
response_by_channel = raw_data.groupby('Sales Channel')['Response'].value_counts(normalize=True).unstack()*100 #normalize=True
ax = response_by_channel.plot(kind='bar', stacked=True)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.legend(title='Response', loc='upper right', labels=['No', 'Yes'])
ax.yaxis.set_major_formatter('{x:.0f}%')
plt.show()
```

```python
raw_data['Response_number'] = raw_data['Response'].map({'No': 0, 'Yes': 1})
raw_data.head()
```

```python
# response_rate = raw_data.groupby('Total Claim Amount')['Response_number'].mean()

# Create a scatter plot
# plt.scatter(raw_data['Total Claim Amount'], raw_data['Response_number'])
# plt.xlabel('Total Claim Amount')
# plt.ylabel('Response Rate')
# plt.title('Response Rate by Total Claim Amount (Scatter Plot)')
# plt.show()

# Define the bin edges for total claim amount
bin_edges = [0, 250, 500, 750, 1000, 1250, 1500, float('inf')]

# Create labels for the bins
bin_labels = ['0-250','251-500', '501-750','751-1000', '1001-1250', '1251-1500', '1500+']

# Cut the 'Total Claim Amount' into bins and create a new column 'Total Claim Amount Bracket'
raw_data['Total Claim Amount Bracket'] = pd.cut(raw_data['Total Claim Amount'], bins=bin_edges, labels=bin_labels)

# Group by 'Total Claim Amount Bracket' and calculate the response rate (mean of 'Response')
response_rate = raw_data.groupby('Total Claim Amount Bracket')['Response'].value_counts(normalize=True).unstack()*100

# Create a bar plot
ax = response_rate.plot(kind='bar', stacked = True)
plt.xlabel('Total Claim Amount Bracket')
plt.ylabel('Response Rate')
plt.title('Response Rate by Total Claim Amount Bracket')
plt.xticks(rotation=45)

ax.yaxis.set_major_formatter('{x:.0f}%')
plt.show()
```

```python
# Using Seaborn to create a KDE plot
sns.kdeplot(data=raw_data, x='Income', hue='Response', fill=True, common_norm=False)
plt.xlabel('Income')
plt.ylabel('Density')
plt.title('Response Rate by Income (KDE Plot)')
plt.show()
```

```python
# Calculate response rate by income bracket

# Define the bin edges for income
bin_edges = [-1, 20000, 40000, 60000, 80000, 100000, float('inf')]

# Create labels for the bins
bin_labels = ['0-20k', '20k-40k', '40k-60k', '60k-80k', '80k-100k', '100k+']

# Cut the 'Income' into bins and create a new column 'Income Bracket'
raw_data['Income Bracket'] = pd.cut(raw_data['Income'], bins=bin_edges, labels=bin_labels)

# raw_data['Income Bracket'].value_counts()
raw_data.head()
```

```python
raw_data['Income Bracket'].value_counts()
```

```python
response_by_income = pd.crosstab(raw_data['Income Bracket'], raw_data['Response']) #, normalize='index'

# response_by_income = raw_data.groupby('Income Bracket')['Response'].value_counts()#.unstack() #normalize=True
response_by_income.plot(kind='bar', stacked=True)
plt.title('Response Rate by Income')
plt.xlabel('Income Bracket')
plt.ylabel('Response Rate')
plt.legend(title='Response', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
